## Exercise 1: Prime Number Checker

```python
num = 17  # Example input
is_prime = True
if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
if is_prime:
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

## Exercise 2: Product of Random Numbers

```python
import random
a = random.randint(1, 10)
b = random.randint(1, 10)
print("Multiply:", a, "×", b)
user_ans = a * b  # simulate user input
print("You entered:", user_ans)
if user_ans == a * b:
    print("Correct!")
else:
    print("Incorrect! The correct answer is", a * b)
```

## Exercise 4: Hospital Billing System

```python
room_charge_per_day = 2000
consultation_fee = 1500
lab_test_cost = 300
days = 4
tests = 3
medicine_charges = 2400
room_total = days * room_charge_per_day
lab_total = tests * lab_test_cost
total_before_discount = room_total + consultation_fee + lab_total + medicine_charges
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0
final_bill = total_before_discount - discount
print("Final Bill Amount: ₹", final_bill)
```
