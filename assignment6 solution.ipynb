## a) 99% Confidence Interval using Sample Standard Deviation

```python
import numpy as np
from scipy.stats import t

data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.2, 1.33, 1.18, 1.22, 1.29]
n = len(data)
mean = np.mean(data)
sd = np.std(data, ddof=1)

alpha = 0.01
t_critical = t.ppf(1 - alpha/2, df=n-1)

lower = mean - t_critical * sd / np.sqrt(n)
upper = mean + t_critical * sd / np.sqrt(n)

(mean, sd, lower, upper)
```

## b) 99% Confidence Interval using Known Population Standard Deviation (Ïƒ = 0.2)

```python
from scipy.stats import norm

sigma = 0.2
z_critical = norm.ppf(1 - 0.01/2)

lower2 = mean - z_critical * sigma / np.sqrt(n)
upper2 = mean + z_critical * sigma / np.sqrt(n)

(mean, lower2, upper2)
```
